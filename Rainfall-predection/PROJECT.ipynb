# For replacing numerical columns 

# Taking mean of the column which has null value and replace with mean

```python
df_mean=df['MinTemp'].mean()
df['MinTemp']=df['MinTemp'].fillna(df_mean)

df_mean=df['MaxTemp'].mean()
df['MaxTemp']=df['MaxTemp'].fillna(df_mean)

df_mean=df['Rainfall'].mean()
df['Rainfall']=df['Rainfall'].fillna(df_mean)

df_mean=df['Evaporation'].mean()
df['Evaporation']=df['Evaporation'].fillna(df_mean)

df_mean=df['Sunshine'].mean()
df['Sunshine']=df['Sunshine'].fillna(df_mean)

df_mean=df['WindGustSpeed'].mean()
df['WindGustSpeed']=df['WindGustSpeed'].fillna(df_mean)

df_mean=df['WindSpeed9am'].mean()
df['WindSpeed9am']=df['WindSpeed9am'].fillna(df_mean)

df_mean=df['WindSpeed3pm'].mean()
df['WindSpeed3pm']=df['WindSpeed3pm'].fillna(df_mean)

df_mean=df['Humidity9am'].mean()
df['Humidity9am']=df['Humidity9am'].fillna(df_mean)

df_mean=df['Humidity3pm'].mean()
df['Humidity3pm']=df['Humidity3pm'].fillna(df_mean)

df_mean=df['Pressure9am'].mean()
df['Pressure9am']=df['Pressure9am'].fillna(df_mean)

df_mean=df['Pressure3pm'].mean()
df['Pressure3pm']=df['Pressure3pm'].fillna(df_mean)

df_mean=df['Cloud9am'].mean()
df['Cloud9am']=df['Cloud9am'].fillna(df_mean)

df_mean=df['Cloud3pm'].mean()
df['Cloud3pm']=df['Cloud3pm'].fillna(df_mean)

df_mean=df['Temp9am'].mean()
df['Temp9am']=df['Temp9am'].fillna(df_mean)

df_mean=df['Temp3pm'].mean()
df['Temp3pm']=df['Temp3pm'].fillna(df_mean)
```

```python
df.isnull().sum()
```

# For replacing object columns

# Taking mode of the column which has null value and replace with mode

```python
mode_value= df["WindGustDir"].mode()[0]
df["WindGustDir"].fillna(mode_value, inplace= True)

mode_value= df["RainToday"].mode()[0]
df["RainToday"].fillna(mode_value, inplace= True)

mode_value= df["WindDir9am"].mode()[0]
df["WindDir9am"].fillna(mode_value, inplace= True)

mode_value= df["WindDir3pm"].mode()[0]
df["WindDir3pm"].fillna(mode_value, inplace= True)
```

# Since the Data is imbalanced , need to do sampling (either Oversampling or Undersampling

```python
import imblearn
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=42)
x_ros, y_ros = ros.fit_resample(x, y)
```

```python
x=x_ros
y=y_ros
```

```python
y.value_counts()     
```

# Importing the models and Train them

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
logreg = LogisticRegression()
knn = KNeighborsClassifier()
dt = DecisionTreeClassifier()
rf = RandomForestClassifier()
```

```python
def mymodel(model):
    model.fit(x_train,y_train)
    y_pred = model.predict(x_test)
    print(classification_report(y_test,y_pred))
    return model
```

# Logistic Regression

```python
mymodel(logreg)
```

```python
# accuracy score on the train data
y_pred = logreg.predict(x_train)
Train_Accuracy_logreg = accuracy_score(y_pred, y_train)
print('Accuracy score of the training data : ', round(Train_Accuracy_logreg*100,2),"%")
```

```python
# accuracy score on the test data
y_pred_on_test_data = logreg.predict(x_test)
Test_Accuracy_logreg = accuracy_score(y_pred_on_test_data, y_test)
print('Accuracy score of the test data : ', round(Test_Accuracy_logreg*100,2),"%")
```

# KNeighborsClassifier

```python
mymodel(knn)
```

```python
# accuracy score on the train data
y_pred = knn.predict(x_train)
Train_Accuracy_knn = accuracy_score(y_pred, y_train)
print('Accuracy score of the training data : ', round(Train_Accuracy_knn*100,2),"%")
```

```python
# accuracy score on the test data
y_pred_on_test_data = knn.predict(x_test)
Test_Accuracy = accuracy_score(y_pred_on_test_data, y_test)
print('Accuracy score of the test data : ', round(Test_Accuracy*100,2),"%")
```

# Decision Tree

```python
mymodel(dt)
```

```python
# accuracy score on the train data
y_pred = dt.predict(x_train)
Train_Accuracy_dt = accuracy_score(y_pred, y_train)
print('Accuracy score of the training data : ', round(Train_Accuracy_dt*100,2),"%")
```

```python
# accuracy score on the test data
y_pred_on_test_data = dt.predict(x_test)
Test_Accuracy_dt = accuracy_score(y_pred_on_test_data, y_test)
print('Accuracy score of the test data : ', round(Test_Accuracy_dt*100,2),"%")
```

# Random forest

```python
mymodel(rf)
```

```python
# accuracy score on the train data
y_pred = rf.predict(x_train)
Train_Accuracy_rf = accuracy_score(y_pred, y_train)
print('Accuracy score of the training data : ', round(Train_Accuracy_rf*100,2),"%")
```

```python
# accuracy score on the test data
y_pred_on_test_data = rf.predict(x_test)
Test_Accuracy_rf = accuracy_score(y_pred_on_test_data, y_test)
print('Accuracy score of the test data : ', round(Test_Accuracy_rf*100,2),"%")
```

```python
ML_Model=['Linear Regression','Decision Tree','Random Forest','k-Nearest Neighbors Regression']

Train_Accuracy=[Train_Accuracy_logreg,Train_Accuracy_dt,Train_Accuracy_rf,Train_Accuracy_knn]
Train_ACC = [round(i,3) for i in Train_Accuracy]

Test_Accuracy=[Test_Accuracy_logreg,Test_Accuracy_dt,Test_Accuracy_rf,Test_Accuracy_logreg]
Test_ACC = [round(i,3) for i in Test_Accuracy]
```

```python
#creating dataframe

Train_ACC_percent = [acc * 100 for acc in Train_ACC]
Test_ACC_percent = [acc * 100 for acc in Test_ACC]

results = pd.DataFrame({ 'ML Model': ML_Model,    
    'Train Accuracy': Train_ACC_percent,
    'Test Accuracy': Test_ACC_percent},index=[1,2,3,4])
```

```python
results
```

```python
# Set up the Seaborn style and color palette
sns.set(style="whitegrid")

# Create the barplot for Test Accuracy only
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='ML Model', y='Test Accuracy', data=results, palette='Set2')

# Adding labels and title
ax.set_xlabel('ML Model')
ax.set_ylabel('Test Accuracy (%)')
ax.set_title('Test Accuracy for Different ML Models')

# Adding the value labels on top of the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}%', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', 
                fontsize=12, color='black', 
                xytext=(0, 5), textcoords='offset points')

# Show the plot
plt.tight_layout()
plt.show()
```
